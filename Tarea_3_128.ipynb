```python
import random
import copy
import matplotlib.pyplot as plt
import numpy as np
from time import time
from memorias import *
import heapq
from multiprocessing.pool import ThreadPool
```

```python
def generarMatrix(archivo):
    lineas = []
    for f in archivo.readlines():
        aux = []
        linea = f.strip(" \r\n").split(" ")
        for i in linea:
            aux.append(int(i))
        lineas.append(aux)
    return lineas
```

```python
def vecindad(inicial, n, tipo):
    while True:
        first = random.randint(0,n-1)
        second = random.randint(0,n-1)
        if(first != second):
            vecino = copy.copy(inicial)
            if(tipo == 1):
                vecino[first] = inicial[second]
                vecino[second] = inicial[first]
                
            elif(tipo == 2):
                minimo = min(first,second)
                maximo = max(first,second)
                aux_max = maximo
                while(minimo <= maximo):
                    vecino[minimo] = inicial[aux_max]
                    minimo +=1
                    aux_max -=1
                    
            elif(tipo == 3):
                minimo = min(first,second)
                maximo = max(first,second)
                vecino[minimo] = inicial[maximo]
                while(minimo <= maximo-1):
                    vecino[minimo+1] = inicial[minimo]
                    minimo +=1
                    
            return vecino
                
```

```python
def funcionObjetivo(actual,n, flujos, distancias,tp):
    costo = 0
    def mult(i):
        c =0
        for j in range(n):
            c += flujos[i][j]*distancias[actual[i]][actual[j]]
        return c
    return sum(tp.map(mult,range(n)))
```

```python
def simulatedAnneling(Tmax, Tmin, iteracionesInternas, tipoEnfriamiento, tipoVecindad, resto = 1, alpha = 0.9):
    tiempo_inicial = time()
    #solucion inicial aleatoria
    inicial = random.sample(range(n),n)
    
    it = 2
    if(tipoEnfriamiento == 3):
        T = Tmax/np.log(it)
    else:
        T = Tmax
    Tmin = Tmin
    costos = []
    mejorCosto = funcionObjetivo(inicial, n, flujos, distancias)
    mejorSolucion = copy.copy(inicial)
    
    while(T > Tmin):

        for i in range(iteracionesInternas):
            s_prima = vecindad(inicial, n, tipoVecindad)
            error = funcionObjetivo(s_prima, n, flujos, distancias) - funcionObjetivo(inicial, n, flujos, distancias)

            if(error < 0):
                inicial = s_prima
                costo = funcionObjetivo(inicial, n, flujos, distancias)
                mejorCosto = costo
                mejorSolucion = copy.copy(inicial)
                
            elif(random.uniform(0, 1) < np.exp(-float(error)/T)):
                inicial = s_prima
                costo = funcionObjetivo(inicial, n, flujos, distancias)

        costos.append(costo)
        it += 1
        TE = ""
        
        if(tipoEnfriamiento == 1): #lineal
            T = T - resto
            TE = "lineal"
            
        elif(tipoEnfriamiento == 2): #geométrico
            T = alpha*T
            TE = "geometrico"
        
        elif(tipoEnfriamiento == 3): #logarítmico
            T = Tmax/np.log(it)
            TE = "logaritmico"
            
    tiempo_final = time() 
    tiempo_ejecucion = tiempo_final - tiempo_inicial
    
    print("Mejor Combinacion: ", mejorSolucion)
    print("Mejor Costo: ", mejorCosto )
    print("Combinacion Actual: ", inicial)
    print("Costo Actual: ", costo)
    plt.plot(costos)
    plt.title("Tmax: "+str(Tmax)+" ,Tmin: "+str(Tmin)+" ,TipoVecindad: "+str(tipoVecindad)+" ,TipoEnfriamiento: "+TE+" ,MejorCosto: "+str(mejorCosto)+", Tiempo Ejecución:"+str(tiempo_ejecucion)+", Tiempo Ejecución minutos:"+str(tiempo_ejecucion/60))
    plt.ylabel("Costos")
    plt.xlabel("Iteracion")
    plt.show()
```
